```python
import pandas as pd
```

```python
df = pd.read_csv('States_Lat_Long.csv')
```

```python
df
```

```python
df1 = pd.read_csv('Lat_Long_data.csv')
```

```python
df1[['Lat', 'Lng']] = df1.latlng.str.split(",", expand = True)
```

```python
df1.drop(columns = ['latlng'], inplace = True)
```

```python
df1 = df1.rename(columns = {'state':'USAState'})
```

```python
df1
```

```python
df_final = df.merge(df1, how = 'outer', on = 'USAState' )
```

```python
df_final.columns
```

```python
df_final = df_final.drop(columns = ['Unnamed: 0','Source'])
```

```python
df_final
```

```python
df_final.to_csv('latlng.csv')
```
